```python
import cv2
import numpy as np

def main():
    # Load your image
    person_img = cv2.imread('my.jpg')

   
    width = int(person_img.shape[1] )
    height = int(person_img.shape[0] )
    dsize = (width, height)
    person_img = cv2.resize(person_img, dsize)

    # Selecting the Region of Interest of the image
    copy_of_person = person_img.copy()

    # Initialize the mask with zeros
    mask = np.zeros(person_img.shape[:2], np.uint8)

    # Define the color range for the foreground (red and green channels)
    min_red_value = 20  # Define your minimum red channel value
    max_red_value = 255  # Define your maximum red channel value
    min_green_value = 10  # Define your minimum green channel value
    max_green_value = 250  # Define your maximum green channel value
    min_blue_value = 10  # Define your minimum blue channel value
    max_blue_value = 255  # Define your maximum blue channel value

    # Create a mask condition based on the color range
    mask_condition = ((person_img[:, :, 0] >= min_red_value) & (person_img[:, :, 0] <= max_red_value) &
                      (person_img[:, :, 1] >= min_green_value) & (person_img[:, :, 1] <= max_green_value)&
                     (person_img[:, :, 2] >= min_blue_value) & (person_img[:, :, 2] <= max_blue_value))

    # Set the mask condition to mark the foreground pixels
    mask[mask_condition] = cv2.GC_PR_FGD

    bgd_model = np.zeros((1, 65), np.float64)
    fgd_model = np.zeros((1, 65), np.float64)

    cv2.grabCut(person_img, mask,None, bgd_model, fgd_model, 100, cv2.GC_INIT_WITH_MASK)
    mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
    no_bg_person = person_img * mask2[:, :, np.newaxis]

    cv2.imshow("Foreground Image", no_bg_person)
    cv2.imwrite('my_image_no_bg.jpeg', no_bg_person)

    # Adding a new Background
    background_img = cv2.imread('historic.jpeg')
    background_img = cv2.resize(background_img, dsize)
    background_img[mask2 != 0] = [0, 0, 0]

    final_img = no_bg_person + background_img
    cv2.imshow("Final Image", final_img)
    cv2.imwrite('final_image.jpeg', final_img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
